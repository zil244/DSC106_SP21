```python
import altair as alt
from vega_datasets import data
import pandas as pd
import numpy as np
```

### Problem 1

```python
df = pd.read_csv('calvinCollegeSeniorScores.csv')
source = df
alt.Chart(source).mark_circle(size=60).encode(
    x='SATM',
    y='SATV',
    size='ACT',
    color='GPA',
    tooltip=['SATM', 'SATV', 'ACT', 'GPA']
).interactive()
```

### Problem 2

```python
brush = alt.selection(type='interval', resolve='global')
base = alt.Chart(source).mark_point().encode(
    color=alt.condition(brush, alt.ColorValue('blue'), alt.ColorValue('gray')),
).add_selection(brush).properties(width=300,height=300)
base.encode(x='SATM',y='SATV') | base.encode(x='ACT',y='GPA')
```
